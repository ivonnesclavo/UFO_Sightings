```python
#Import Dependencies
import pandas as pd
```

```python
#Path to CSV
original_ufo_csv = "./ufo-sightings-transformed.csv"
```

```python
#Clean Dataset
original_ufo_df = pd.read_csv(original_ufo_csv)
original_ufo_df.head()
```

```python
#Take only useful columns
ufo_df = original_ufo_df[['Date_time','Year','Month','Hour','Season','Country','Region','Locale','latitude','longitude','UFO_shape','length_of_encounter_seconds']]
ufo_df.head()
```

```python
#Renameing
ufo_df.rename(columns={'length_of_encounter_seconds':'Duration_seconds'},inplace=True)
ufo_df.head()
```

```python
#Filter only United States
us_ufo_df = ufo_df[ufo_df['Country']== 'United States']
us_ufo_df.head()
```

```python
#Drop "nan" values for shape
us_ufo_df['UFO_shape'].unique()
```

```python
us_ufo_df.dropna(subset=['UFO_shape'], inplace=True)
us_ufo_df['UFO_shape'].unique()
```

```python
#Export dataframe
us_ufo_df.to_csv("./us_ufo.csv",index=False)
```
